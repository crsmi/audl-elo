With the new games format it is probably better to have consistent team abbrs in the resuts input file (audl_games_w_neutral.csv)

Lets put together a mapping right now.

```python
teams = pd.read_csv('audl_teams.csv')
frans = pd.read_csv('audl_franchises.csv',index_col='franchise')

#Create list of teams for each franchise
fran_teams = {}
#Create mapping of team abbr to team abbr through franchise
newest_team_abbr_map = {}

for row in teams.itertuples():
    if row.franchise in fran_teams:
        fran_teams[row.franchise].append(row.abbr)
    else:
        fran_teams[row.franchise] = [row.abbr]
        
    newest_team_abbr = frans.loc[row.franchise]['abbr']    
    newest_team_abbr_map[row.abbr] = newest_team_abbr
        
newest_team_abbr_map
```

Lets apply this to the file quick.

```python
games = pd.read_csv('rating/audl_games_w_neutral.csv')
games['team1'] = games['team1'].map(newest_team_abbr_map)
games['team2'] = games['team2'].map(newest_team_abbr_map)
games.to_csv('rating/audl_games_w_neutral_and_correct_team.csv',index=False)
```
